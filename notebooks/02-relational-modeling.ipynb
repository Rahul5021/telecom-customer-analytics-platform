```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/cleaned_churn.csv")

df.head()
```

```python
df = df.reset_index(drop=True)
df['customer_id'] = df.index + 1
```

```python
df[['customer_id']].head()
```

```python
customers = df[['customer_id', 'State', 'Account length', 'Churn']].copy()
customers.columns = ['customer_id', 'state', 'account_length', 'churn']
```

```python
subscriptions = df[['customer_id', 'International plan', 'Voice mail plan']].copy()
subscriptions.columns = ['customer_id', 'international_plan', 'voice_mail_plan']
```

```python
usage_metrics = df[[
    'customer_id',
    'Total day minutes', 'Total eve minutes',
    'Total night minutes', 'Total intl minutes',
    'Total day calls', 'Total eve calls',
    'Total night calls', 'Total intl calls'
]].copy()

usage_metrics.columns = [
    'customer_id',
    'day_minutes', 'eve_minutes',
    'night_minutes', 'intl_minutes',
    'day_calls', 'eve_calls',
    'night_calls', 'intl_calls'
]
```

```python
support_metrics = df[['customer_id', 'Customer service calls']].copy()
support_metrics.columns = ['customer_id', 'customer_service_calls']
```

```python
df['estimated_revenue'] = (
    df['Total day minutes'] * 0.17 +
    df['Total eve minutes'] * 0.085 +
    df['Total night minutes'] * 0.045 +
    df['Total intl minutes'] * 0.27
)

revenue_summary = df[['customer_id', 'estimated_revenue']].copy()

revenue_summary['revenue_risk_flag'] = np.where(
    df['Churn'] == 1,
    'At Risk',
    'Stable'
)
```

```python
customers.to_csv("../data/relational/customers.csv", index=False)
subscriptions.to_csv("../data/relational/subscriptions.csv", index=False)
usage_metrics.to_csv("../data/relational/usage_metrics.csv", index=False)
support_metrics.to_csv("../data/relational/support_metrics.csv", index=False)
revenue_summary.to_csv("../data/relational/revenue_summary.csv", index=False)
```
